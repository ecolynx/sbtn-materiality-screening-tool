```python
import pandas as pd
from json import loads, dumps
```

# Definition of pressure categories (from ENCORE)

```python
df = pd.read_excel('docs/SBTN-Materiality-Screening-Tool-v1.xlsx', 
                   engine='openpyxl', 
                   sheet_name="3 - Methodology", 
                   usecols="O:Q", 
                   skiprows=4, 
                   nrows=12).fillna(method='ffill')
```

```python
df
```

```python
df.groupby('Nature-related issue area').apply(lambda x: x)
```

```python
result = df.to_json(orient="records")
parsed = loads(result)
with open('encore-pressure-categories.json', 'w') as outfile:
    outfile.write(dumps(parsed, indent=4))
```
